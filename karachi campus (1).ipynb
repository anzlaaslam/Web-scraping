```python
import requests
# to scrap data
```

```python
from bs4 import BeautifulSoup
# to extract html content
```

```python
URL = "https://khi.nu.edu.pk/faculty-php/"
r = requests.get(URL)
print(r)
# returns no of requests
```

```python
soup = BeautifulSoup (r.content, 'html.parser')
```

```python
soup.prettify
```

```python
#f = soup.findAll('span', class_='kingster-mm-menu-blank')
f = soup.findAll('a')
```

```python
len(f)
```

```python
#a=soup.findAll('div',{"id":"faculty-tab"})
```

```python
#a=soup.findAll('div',class_ = "gdlr-core-hover-opacity").text
name = soup.find('h3', class_='gdlr-core-personnel-list-title').text.split(',')[0]
name
```

```python
image = soup.find('div', class_='gdlr-core-personnel-list-image').find('img')['src']
image
```

```python
designation = soup.find('div', class_='gdlr-core-personnel-list-position').text
designation
```

```python
email = soup.find('span', class_='gdlr-core-icon-list-content').text
email
```

```python
extension = soup.findAll('span', class_='gdlr-core-icon-list-content')[1].text
#soup.find('div', class_='kingster-personnel-info-list kingster-type-phone')
extension
```

```python
name_=[]
image_=[]
designation_=[]
hec_=[]
extension_=[]
education_=[]
email_=[]
```

```python
print(len(image_))
print(len(name_))
print(len(designation_))
print(len(hec_))
print(len(extension_))
print(len(education_))
print(len(email_))
```

```python
for i in range(len(f)):
    faculty = f[i]
    #f = soup.findAll('a')[i]
    try:
        name_.append(faculty.find('div', class_='gdlr-core-title-item-title-wrap').text.split(',')[0])
    except:
        name_.append('None')
    try:
        image_.append("https://khi.nu.edu.pk/faculty-php/"+faculty.find('div', class_='gdlr-core-personnel-list-image').find('img')['src'])
    except:
        image_.append('None')
```

```python
print(len(image_))
print(len(name_))
```
